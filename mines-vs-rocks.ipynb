```python
df = pd.read_csv('/kaggle/input/mines-vs-rocks/sonar.all-data.csv', header=None)
df
```

```python
data = df.to_numpy()
X, Y = data[:,:59], data[:,60]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.12, stratify=Y) 
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
print(f'Accuracy on training data is {model.score(x_train, y_train)}')
print(f'Accuracy on test data is {model.score(x_test, y_test)}')
```
